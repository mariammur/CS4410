```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import requests
from textblob import TextBlob
```

```python
response = requests.get('https://www.python.org')
response.content # gives back the page's HTML
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html5lib')
text = soup.get_text(strip=True) # text without tags
```

```python
blob=TextBlob(text)
blob
```

```python
blob.sentences
```

```python
full_text_word = ' '
stopwords = set(STOPWORDS)
```

```python
text_list = text.split()

for i in range(len(text_list)):

  text_list[i] = text_list[i].lower()
```

```python
for words in text_list:
  full_text_word = full_text_word + words + ' '
```

```python
wordcloud = WordCloud(width = 800, height = 800,background_color ='white',stopwords = stopwords,min_font_size = 10).generate(full_text_word)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.savefig('wordcloud.png')
plt.show()
```
